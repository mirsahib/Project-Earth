```julia
wikiData = readdlm("wikipediaEVDdatesformated.csv",',');# loading data set
```

```julia
wikiData[end-9:end,:] # slicing last 10 row
```

<font color=red>As you can see some data is missing in the data set this is beacuse when the disease breakout the countries doesn't report the initial cases,
We need to formate those column so that there has no empty cell</font><br>
**size()** function is returning the number of rows and columns in the dataset<br>**isnumber()** function is parsing the string to number

```julia
rows,cols = size(wikiData)
for j= 1:cols #traversing column
    for i=1:rows #traversing the row
        if !isnumber(string(wikiData[i,j]))#if cell is empty
            wikiData[i,j]=0 #replace it with 0
        end
    end
end
```

<font color=red>Now you can see that the missing cell is replaced by zero</font>

```julia
wikiData[end-9:end,:]# slicing the last 10 row
```

**epiday** stores the number of days since the disease occur i.e 22 Mar 2014 <br>
**EVDcasesbycountry** stores the number of cases by country (Guinea,Liberia,Sierra Leone)

```julia
epiday = wikiData[:,1]
EVDcasesbycountry = wikiData[:,[4,6,8]]
```

```julia
using Plots
pyplot()
plot(epiday,EVDcasesbycountry,
legend = :topleft,
label = ["Guinea" "Liberia" "Sierra Leone"],
title = "EVD in West Africa, epidemic segregated by country",
xlabel = "Days since 22 Mar 2014",
ylabel = "Number of cases to date",
line = :path
)
```
